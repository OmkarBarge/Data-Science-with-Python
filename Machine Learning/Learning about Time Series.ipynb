### Time Series 
A time series is a collection of observations of well defined data items obtained through repeated measurements over time. <br>

Examples:<br>
1) Retail sales each month of the year<br>
2) Stock price data etc


Time series data is a sequence of data points indexed in time.

### Terminologies

1)<b> Trend</b><br>
Movement of data values up or down with time for a particular time span.

2) <b>Seasonality</b><br>
Periodic flutuations with time

3) <b>Stationary</b><br>
Time series data must be stationary<br>
    
    1) Data should have constant mean and varaince with respect to time
    2) Minimal trend and minimal seasonality

4) <b>Noise/Residual/Irregular components</b><br>
The portion of time series data that cant be explaiend by trend and/or seasonality is noise or irregular components.

#### Time Series Data Decomposition

a) Additive approach -> y(t) = Trend + Seasonality + Residuals<br>
b) Multiplicative Approach -> y(t) = Trend * Seasonality * Residuals<br>


#### Time Series data must bear these properties
1) There should have no null values<br>
2) Data must be sequentially indexed with time<br>
3) Data must have autocorrelation. (Present time series values must be dependent on previous time series values)

#### How to forecast time series data?

### ARIMA (Auto Regressive Integrated Moving Average)
It operates on 3 terms - p,d,q

<b>p</b> - Number of Auto Regressive terms. Determined using the PACF(Partial Auto Correlation Function) plot

<b>q</b> - Number of lagged forecast errors in the prediction equation. Determined using the ACF(Auto Correlation Function) plot
    
<b>d</b> - Order of differencing.     

ARIMA can be imported from statsmodels library

#### How to Evaluate the performance of the ARIMA model?
ARIMA model results can be evaluated using<br>
1) Mean Squared Error<br>
3) Root Mean Sqaured Error<br>
2) Mean Absolute Error

### Auto_ARIMA
1) It automatically generates the optimal values of p,d and q based on min and max value of p, d and q that the user sets as the model hyperparamters<br>
2) Auto ARIMA can be imported from pmd_arima<br>
