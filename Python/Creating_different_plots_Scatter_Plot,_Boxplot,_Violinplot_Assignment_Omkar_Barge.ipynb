```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/healthexp.csv')
```

1. Create a Scatter plot of 'Year wise Life Expectancy based on
Countries' and apply the Following customization to the plot.<br>
● Make a plot by setting figure size (10.6) using seaborn
set() function.<br>
● Set the color palette for the plot using the set_palette()
function as "tab10".<br>
● Set the style for the plot using the set_style() function as
"darkgrid".<br>
● Set the context for the plot using set_cotext() function
with following customization ("paper", font_scale = 2,
rc={"grid.linewidth": 0.5}).<br>
● Make a Bar plot using seaborn scatterplot() function
setting
x='Country',y='Life_Expectancy',hue='Country',style='Count
ry',size = 'Country'.<br>
● Add the title to the plot as 'Year wise Life Expectancy
based on Countries'.<br>
● Add X-label as 'Year' and Y-label as 'Life Expectancy over a
year'.<br>
● Anchor the legend using matplot function as :-
plt.legend(bbox_to_anchor=(1, 1))<br>
● Add grid to the plot using matplotlib grid() function.<br>
● Show the plot using the show() function of matplotlib.

```python
sns.set(rc={'figure.figsize':(10,6)})
sns.set_palette("tab10")
sns.set_style('darkgrid')
sns.set_context("paper", font_scale = 2, rc={"grid.linewidth": 0.5})
sns.scatterplot(data=df,x='Year',y='Life_Expectancy',hue='Country',style='Country',size = 'Country')
plt.title('Year wise Life Expectancy based on Countries')
plt.xlabel('Year') 
plt.ylabel('Life Expectancy over a year')
plt.legend(bbox_to_anchor=(1, 1))
plt.grid() 
plt.show()
```

2. Create a Boxplot of Country wise Spending in USD - Boxplot'
and apply the Following customization to the plot.<br>
● Make a plot by setting figure size (10.6) using
seaborn set() function.<br>
● Set the color palette for the plot using the
set_palette() function as "hls".<br>
● Set the style for the plot using the set_style()
function as "darkgrid".<br>
● Set the context for the plot using set_cotext()
function with following customization ("paper",
font_scale = 2, rc={"grid.linewidth": 0.5}).<br>
● Make a Box plot using seaborn boxplot() function
setting x='Country',y='Spending_USD'.<br>
● Add the title to the plot as Country wise Spending in
USD - Boxplot'.<br>
● Add X-label as 'Country' and Y-label as
'Spending(USD) on Healthcare'.<br>
● Add grid to the plot using matplotlib grid() function.
<br>
● Show the plot using the show() function of
matplotlib.

```python
sns.set(rc={'figure.figsize':(10,6)})
sns.set_palette("hls")
sns.set_style('darkgrid')
sns.set_context("paper", font_scale = 2, rc={"grid.linewidth": 0.5})
sns.boxplot(data=df,x='Country',y='Spending_USD')
plt.title('Country wise Spending in USD - Boxplot')
plt.xlabel('Country')
plt.ylabel('Spending(USD) on Healthcare')
plt.grid()
plt.show() 
```


3. Create a Violin plot of 'Country wise Life Expectancy - Violin
plot' and apply the Following customization to the plot.<br>
● Make a plot by setting figure size (10.6) using
seaborn set() function.<br>
● Set the color palette for the plot using the
set_palette() function as "husl".<br>
● Set the style for the plot using the set_style()
function as "darkgrid".<br>
● Set the context for the plot using set_cotext()
function with following customization ("paper",
font_scale = 2, rc={"grid.linewidth": 0.5}).<br>
● Make a Violin plot using seaborn violinplot()
function setting x='Country',y='Life_Expectancy'.<br>
● Add the title to the plot as 'Country wise Life
Expectancy - Violin plot'.<br>
● Add X-label as 'Country' and Y-label as 'Life
Expectancy'.<br>
● Show the plot using the show() function of
matplotlib.

```python
sns.set(rc={'figure.figsize':(10,6)})
sns.set_palette("husl")
sns.set_style('darkgrid')
sns.set_context("paper", font_scale = 2, rc={"grid.linewidth": 0.5})
sns.violinplot(data=df,x='Country',y='Life_Expectancy')
plt.title('Country wise Life Expectancy - Violin plot')
plt.xlabel('Country') 
plt.ylabel('Life Expectancy')
plt.show() 
```
