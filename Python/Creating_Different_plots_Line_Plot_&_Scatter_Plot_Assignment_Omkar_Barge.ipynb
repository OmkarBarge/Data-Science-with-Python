```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/DataScience_Marks_Dataset.csv')
```

# 1.Create a simple line plot of 'Name'vs'Total_marks'.

```python
plt.plot(df['Name'],df['Total_marks'])
```

## 2.Create a line plot of 'Name'vs'SQL’ and apply the Following customization to the plot.<br>●Make the Figure size(9,4)<br>●Make line/markers color 'indigo'<br>●Make line width as 2 and line style as'-.'<br>●Add label as python<br>●Add markers as 'o' with marker size = 10 <br>●Add a title to the Plot with fontsize=15 and color=purple<br>●Add X-Label as Name &Y - Label as Python<br>●Add Grid to the plot using grid() function<br>●Show the legend for the plot using legend() funct<br>●Also Show the plot1


```python
plt.figure(figsize=(9,4))
plt.plot(df['Name'],df['SQL'],marker='o',markersize=10,lw =2,linestyle='-',color='indigo',label='Python')
plt.title('SQL Marks',fontsize=10)
plt.xlabel('Name')
plt.ylabel('Python')
plt.grid()
plt.legend()
plt.show()
```

### 3.Create a Line plot of 'Name' vs 'Python' , 'Statistics' and 'SQL' and apply the Following customization to the plot.<br>●Make the Figure size(9,4)<br>●Make the line of 'Name' vs 'Python' with color='cyan',marker='^', label='Python',marker size=10<br>●Make the line of 'Name' vs 'Statistics' with color='red',marker='o',label='Statistics',marker size =10<br>●Make the line of 'Name' vs 'SQL' with color = 'purple',marker='*',label='SQL',marker size=10<br>●Add a title to the Plot with font size =1 3 and color ='Blue'<br>●Add X-Label as Name<br>●Add Grid to the plot using grid() function<br>●Show the legend for the plot using legend() function<br>●Also Show the plot

```python
plt.figure(figsize=(9,4))
plt.plot(df['Name'],df['Python'],color='cyan',marker='^',label='Python',markersize=10)
plt.plot(df['Name'],df['Statistics'],color='red',marker='o',label='Statistics',markersize=10)
plt.plot(df['Name'],df['SQL'],color='purple',marker='*',label='SQL',markersize=10)
plt.title('Marks of Three Subjects',fontsize=13,color='blue')
plt.xlabel('Name')
plt.grid()
plt.legend()
plt.show()
```

## 4.Create a Scatter plot of Python vs Total marks and apply the Following customization to the plot.<br>●Make the Figuresize(9,4)<br>●Make the scatter plot with marker='h',s='40'and color='Orange'<br>●Add X-Label as Python &Y-Label as Total_marks<br>●Add a title to the Plot with font size=15 and color='purple'<br>●Add Grid to the plot using grid() function ● Also Show the plot

```python
plt.figure(figsize=(9,4))
plt.scatter(df['Name'],df['Total_marks'],marker='h',s=40,color='orange')
plt.xlabel('Python')
plt.ylabel('Total Marks')
plt.title('Scatter Plot',fontsize=15,color='purple')
plt.grid()
plt.show()
```

## 5.Create a Scatter plot of Name vs Python,Statistics and SQL' in the same plot and apply the Following customization to the plot.<br>●Make the Figuresize(9,4)<br>●Make a scatter plot of 'Name' vs 'Python' with color='teal',marker='s',s=45,label='Name vs Python'<br>●Make a scatter plot of 'Name' vs 'Statistics' with color = 'salmon',marker='p',s=45,label= 'Name vs Statistics'<br>●Make a scatter plot of 'Name' vs 'SQL' with color = 'orchid',marker='h',s=45,label='Name vs SQL'<br>●Add a title to the Plot with fontsize = 15 and color = 'black'<br>●Add X-Lable as Name<br>●Add Grid to the plot using grid() function<br>●Show the legend for the plot using legend()function<br>●Also Show the plot.

```python
plt.figure(figsize=(9,4))
plt.scatter(df['Name'],df['Python'],color='teal',marker='s',s=45,label='Name vs Python')
plt.scatter(df['Name'],df['Statistics'],color='salmon',marker='p',s=45,label='Name vs Statistics')
plt.scatter(df['Name'],df['SQL'],color='orchid',marker='h',s=45,label='Name vs SQL')

plt.title('Scatter Plot',fontsize=15,color='black')
plt.xlabel('Name')
plt.grid()
plt.legend()
plt.show()
```
