```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/DataScience_Marks_Dataset.csv')
```

## 1.Create a simple Histogram for ‘Total_marks’

```python
plt.hist(df['Total_marks'])
plt.show()
```

##2.Create a Histogram for 'Total_marks' and apply the Following customization to the plot.<br>●Make the Figuresize(9,4)<br>●Make Histogram color = 'teal'<br>●Make Edge color as 'black'<br>●Define bins for Histogram as bins = 10<br>●Add a title to the Plot with fontsize=15 and color = purple<br>●Add X-Label as 'TotalMarks'<br>●Also Show the plot.


```python
plt.figure(figsize=(9,4))
plt.hist(df['Total_marks'],color='teal',edgecolor='black',bins=10)
plt.title('Histogram',fontsize=15,color='purple')
plt.xlabel('TotalMarks')
plt.show()
```

## 3.Create a Piechart of 'Name wise Python Marks' and apply the Following customization to the plot.<br>●Make the Figuresize(5,5)<br>●Define pie plot with x as 'Python' marks and label as 'Name'<br>●Pass the color list for color of piechart sections as colors=['red','cornflowerblue','yellow','tan','brown','green','cyan','pink']<br>●Define autopact till two decimal points as autopct='%.2f%%'<br>●Make shadow = True<br>●Make Counter clock wise = False<br>●Define color of the labels using txtprops as 'navy'color.<br>●Add a title to the Plot with fontsize=15 and color='teal'<br>●Also Show the plot

```python
plt.figure(figsize=(5,5))
plt.pie(df['Python'],labels=df['Name'],colors=['red','cornflowerblue','yellow','tan','brown','green','cyan','pink'],autopct='%.2f%%',shadow=True,counterclock=False,textprops=dict(color='navy'))
plt.title('Pie Plot',fontsize=15,color='teal')
plt.show()
```

## 4.Create a Pie chart of 'Name wise Total Marks' and apply the Following customization to theplot.<br>●Make the Figuresize(5,5)<br>●Define pie plot with x as 'Total_marks' and label as 'Name'<br>●Pass the color list for color of pie chart sections as colors=['red','blue','yellow','green','cyan','deeppink','dodgerblue','orange']<br>●Define autopact till two decimal points as autopct='%.2f%%'<br>●Make the piechart explode by defining explode=(0,0,0,0,0,0.2,0,0.1)<br>●Make shadow = True <br>●Make Counterclockwise=False<br>●Define legend()with legend content as Name and title='Names'<br>●Position the legend by defining loc='right'and bbox_to_anchor=(1,1,0.6,-0.5)<br>●Add a title to the Plot with fontsize=15 and color='blue'.


```python
plt.figure(figsize=(5,5))
plt.pie(df['Total_marks'],labels=df['Name'],colors=['red','blue','yellow','green','cyan','deeppink','dodgerblue','orange'],autopct='%.2f%%',explode=(0,0,0,0,0,0.2,0,0.1),shadow=True,counterclock=False)
plt.legend(['Name'],loc='right',bbox_to_anchor=(1,1,0.6,-0.5))
plt.title('Pie Plot',fontsize=15,color='blue')
plt.show()
```
