```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/healthexp.csv')
```

1. Create a KDE plot for Spending on Healthcare (USD) based on
Countries' and apply the Following customization to the plot.<br>
● Make a plot by setting figure size (15,6) using seaborn
set() function.<br>
● Set the color palette for the plot using the set_palette()
function as "tab10".<br>
● Set the style for the plot using the set_style() function as
"darkgrid".<br>
● Set the context for the plot using set_cotext() function
with following customization ("paper", font_scale = 2,
rc={"grid.linewidth": 0.5}).<br>
● Make a KDEplot using seaborn kdeplot() function setting
x='Spending_USD',hue='Country',palette='tab10'.<br>
● Add the title to the plot as 'KDE plot for Spending on
Healthcare (USD) based on Countries'.<br>
● Show the plot using the show() function of matplotlib.

```python
sns.set(rc={'figure.figsize':(15,6)})
sns.set_palette("tab10")
sns.set_style('darkgrid')
sns.set_context("paper", font_scale = 2, rc={"grid.linewidth": 0.5})
sns.kdeplot(data=df,x='Spending_USD',hue='Country',palette='tab10')
plt.title('KDE plot for Spending on Healthcare (USD) based on Countries')
plt.show()
```

2. Create a Histogram of 'Histogram for Life Expectancy based on
Year (2019,2020,2021)' and apply the Following customization to the
plot.<br>
● Make a new data frame having Year > 2018 and use it to
plot a histogram.<br>
● Make a plot by setting figure size (15,8) using seaborn
set() function.<br>
● Set the color palette for the plot using the set_palette()
function as "hls".<br>
● Set the style for the plot using the set_style() function as
"darkgrid".<br>
● Set the context for the plot using set_cotext() function
with following customization ("paper", font_scale = 2,
rc={"grid.linewidth": 0.5}).<br>
● Make a Histogram plot using seaborn histplot() function
setting x='Life_Expectancy',hue='Year',palette='bright'.<br>
● Add the title to the plot as 'Histogram for Life Expectancy
based on Year (2019,2020,2021)'.<br>
● Show the plot using the show() function of matplotlib.

```python
df_2018=df[df["Year"] > 2018]
df_2018.head()
sns.set(rc={'figure.figsize':(15,8)})
sns.set_palette("hls")
sns.set_style('darkgrid')
sns.set_context("paper", font_scale = 2, rc={"grid.linewidth": 0.5})
sns.histplot(data=df_2018,x='Life_Expectancy',hue='Year',palette='bright')
plt.title('Histogram for Life Expectancy based on Year (2019,2020,2021)')
plt.show()
```

3. Create a Heatmap and apply the Following customization to the
plot.<br>
● Make a correlation table by using corr() function.<br>
● Set figure size as (8,4) using matplotlib figure() function
as plt.figure(figsize=(8,4))<br>
● Make a Heatmap using seaborn heatmap() function
setting data=corr,annot=True,cmap='RdBu'.<br>
● Show the plot using the show() function of matplotlib.

```python
corr = df.corr()
corr
plt.figure(figsize=(8,4))
sns.heatmap(data=corr,annot=True,cmap='RdBu')
plt.show()
```

4. Create a Jointplot and apply the Following customization to the
plot.<br>
● Set the style for the plot using the set_style() function as
"darkgrid".<br>
● Set the context for the plot using set_cotext() function
with following customization ("paper", font_scale = 2,
rc={"grid.linewidth": 0.5}).<br>
● Make a Jointplot using seaborn jointplot() function
setting
x='Spending_USD',y='Year',hue='Country',palette='bright',ki
nd='kde',height=10.<br>
● Show the plot using the show() function of matplotlib.

```python
sns.set_style('darkgrid')
sns.set_context("paper", font_scale = 2, rc={"grid.linewidth": 0.5})
sns.jointplot(data=df,x='Spending_USD',y='Year',hue='Country',palette='bright',kind='kde',height=10)
plt.show()
```
