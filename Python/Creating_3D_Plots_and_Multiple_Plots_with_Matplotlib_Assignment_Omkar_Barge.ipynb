>> ### ➔Plot-3<br>Create a Pie chart of 'Name wise Total Marks' and apply the Following customization to the plot.<br>◆Define pie plot with x as'Total_marks' and label as 'Name'<br>◆Pass the color list for color of pie chart sections as colors=['red','blue','yellow','green','cyan','deeppink','dodgerblue','orange']<br>◆Define autopact till two decimal points as autopct='%.2f%%'<br>◆Make the pie chart explode by defining explode=(0,0,0,0,0,0.2,0,0.1)<br>◆Make shadow = True<br>◆Make Counterclockwise = False<br>◆Define legend() with legend content as Name and title = 'Names'<br>◆Position the legend by defining loc = 'right' and bbox_to_anchor=(0.9,1,0.6,-0.5)<br>◆Add a title to the Plot with font size =15 and color='blue'.

>>> ### ➔Plot-4)<br>Create a Stack-Area chart of 'Name wise Total Python,SQL and Statistics marks' and apply the following customization to the plot.<br>◆Define Stack-Area plot for Name wise Total Python,SQL and Statistics marks with labels for Python,SQL,Statistics.<br>◆Set X-Label as 'Name'<br>◆Set Y-Label as 'Marks'<br>◆Set the limit for Y-Label as (0,400)<br>◆Show the legend for the Plot using legend() function <br>◆Add a title to the Plot using title()function with fontsize=15 and color='blue'

```python
fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2,figsize=(10,12))

# ax1.set_figure(figsize=(10,8))
ax1.bar(df['Name'],df['SQL'],color='orange',edgecolor='black')
ax1.set_title('Bar Plot',fontsize=15,color='blue')
ax1.set_xlabel('Name')
ax1.set_ylabel('SQL')

ax2.plot(df['Name'],df['Python'],color='cyan',marker='^',label='Python',markersize=10)
ax2.plot(df['Name'],df['Statistics'],color='red',marker='o',label='Statistics',markersize=10)
ax2.plot(df['Name'],df['SQL'],color='purple',marker='*',label='SQL',markersize=10)
ax2.set_title('Line Plot',fontsize=15,color='blue')
ax2.set_xlabel('Name')
ax2.set_ylabel('Marks')
ax2.grid()
ax2.legend()

ax3.pie(df['Total_marks'],labels=df['Name'], colors=['red','blue','yellow','green','cyan','deeppink','dodgerblue','orange'],autopct='%.2f%%',explode=(0,0,0,0,0,0.2,0,0.1),shadow=True,counterclock=False)
ax2.legend(['Names'],loc='right',bbox_to_anchor=(0.9,1,0.6,-0.5))
ax3.set_title('Pie Plot',fontsize=15,color='blue')

ax4.stackplot(df['Name'],df['Python'],df['SQL'],df['Statistics'],labels=['Python','SQL','Statistics'])
ax4.set_xlabel('Names')
ax4.set_ylabel('Marks')
ax4.set_ylim(0,400)
ax4.legend()
ax4.set_title('StackPlot',fontsize=15,color='blue')
```

### 2.Create a 3D scatter plot of Python,SQL and Statistics marks' and apply the following customization to the plot.<br>●Define Scatter plot for Python,SQL and Statistics<br>●Set x-Label as'Python'<br>●Set y-Label as 'SQL'<br>●Set z-Label as 'Statistics'<br>●Rotate the X-label with 90 degrees<br>●Add a title to the Plot using set_title()function with fontsize=15 and color='blue'<br>●Also Show the plot.

```python
from mpl_toolkits.mplot3d import Axes3D
fig=plt.figure()
ax=plt.axes(projection='3d')
```

```python
fig = plt.figure(figsize=(12,9))
ax = fig.add_subplot(projection='3d')
ax.scatter(df['Python'],df['Statistics'],df['SQL'])
ax.set_xlabel('Python',rotation=90)
ax.set_ylabel('Statistics')
ax.set_zlabel('SQL')
ax.set_title('3D Plot',fontsize=15,color='blue')
```
