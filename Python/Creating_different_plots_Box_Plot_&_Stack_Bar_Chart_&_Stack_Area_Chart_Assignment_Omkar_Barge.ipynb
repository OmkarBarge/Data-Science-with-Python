```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/DataScience_Marks_Dataset.csv")
```

## 1.Create a simple Box Plot for Total_marks.

```python
plt.boxplot(df['Total_marks'])
plt.show()
```

## 2.Create a Stack - Bar Plot using matplotlib bar() function,for 'Python' and 'SQL and apply the Following customization to the plot.<br>●Make the Figuresize(5,5)<br>●Customize the bars of 'Python' with color = 'teal' and label='Python'<br>●Customize the bars of 'SQL' with color='Orange',label='SQL'and define bottom as SQL<br>●Add X-Label as 'Name'<br>●Show the legend for the Plot using legend() function<br>●Add a title to the Plot using title() function<br>●Rotate the X-label with 90 degrees.<br>●Show the grip for the plot using grid() function.<br>●Set the limit for Y-Label as (0,250)<br>●Also Show the plot

```python
plt.figure(figsize=(5,5))
plt.bar(df['Name'],df['Python'],color='teal',label='Python')
plt.bar(df['Name'],df['SQL'],color='orange',label='SQL')
plt.xlabel("Name",rotation=90)
plt.legend()
plt.title('Bar Plot')
plt.grid()
plt.ylim(0,250)
```

## 3.Create a Stack-Bar Plot using pandas plot() function,for'Python'and'SQL'using plot kind='bar'and stacked='True'also show the plot with grids

```python
df['Python'].plot(kind='bar',stacked=True)
plt.grid()
```

```python
df['SQL'].plot(kind='bar',stacked=True)
plt.grid()
```

## 4.Create a Stack-Area chart of 'Name wise Total Python,SQL and Statistics marks' and apply the following customization to the plot.<br>●Make the Figure size(10,6)<br>●Define Stack-Area plot for Name wise Total Python,SQL and Statistics marks with labels for Python,SQL,Statistics.<br>●Set X-Label as 'Name'<br>●Rotate the X-label with 90 degrees<br>●Set the limit for Y-Label as (0,400)<br>●Show the legend for the Plot using legend()function<br>●Add a title to the Plot using title() function with fontsize=15 and color='blue'<br>●Also Show the plot.

```python
plt.figure(figsize=(10,6))
plt.stackplot(df['Name'],df['Python'],df['SQL'],df['Statistics'],labels=['Python','SQL','Statistics'])
plt.xlabel('Name',rotation=90)
plt.ylim(0,400)
plt.legend()
plt.title('Stack Plot',fontsize=15,color='blue')
plt.show()
```
