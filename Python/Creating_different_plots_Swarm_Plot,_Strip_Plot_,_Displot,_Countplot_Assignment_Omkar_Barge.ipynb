```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('/content/healthexp.csv')
```

```python
df.head()
```

> # 1. 
    Create a Strip plot of 'Country wise Spending in USD - Strip
    plot' and apply the Following customization to the plot.
    ● Make a plot by setting figure size (10.6) using seaborn
    set() function.
    ● Set the color palette for the plot using the set_palette()
    function as "tab10".
    ● Set the style for the plot using the set_style() function as
    "darkgrid".
    ● Set the context for the plot using set_cotext() function
    with following customization ("paper", font_scale = 2,
    rc={"grid.linewidth": 0.5}).
    ● Make a Strip plot using seaborn stripplot() function
    setting x='Country',y='Spending_USD'.
    ● Add the title to the plot as 'Country wise Spending in USD
    - Strip plot'.
    ● Add X-label as 'Country' and Y-label as 'Spending(USD) on
    Healthcare'.
    ● Add grid to the plot using matplotlib grid() function.
    ● Show the plot using the show() function of matplotlib.


```python
sns.set(rc={'figure.figsize':(10,6)})
sns.set_palette('tab10')
sns.set_style('darkgrid')
sns.set_context('paper', font_scale = 2,rc={'grid.linewidth': 0.5})
sns.stripplot(data=df,x = 'Country', y='Spending_USD')
plt.title('Country wise Spending in USD')
plt.xlabel('Country')
plt.ylabel('Spending(USD)')
plt.grid()
plt.show()
```

> # 2. 
    Create a Swarm plot of 'Country wise Life Expectancy based
    on Year' and apply the Following customization to the plot.
    ● Make a plot by setting figure size (10.6) using seaborn
    set() function.
    ● Set the color palette for the plot using the set_palette()
    function as "hls".
    ● Set the style for the plot using the set_style() function as
    "darkgrid".
    ● Set the context for the plot using set_cotext() function
    with following customization ("paper", font_scale = 2,
    rc={"grid.linewidth": 0.5}).
    ● Make a Swarm plot using seaborn swarmplot() function
    setting x='Country',y='Life_Expectancy',hue='Year'.
    ● Add the title to the plot as 'Country wise Life Expectancy
    based on Year'.
    ● Add X-label as 'Country' and Y-label as 'Country wise Life
    Expectancy based on year'.
    ● Add legend and apply following methods ncol=2,
    fontsize=12,markerscale=0.8, bbox_to_anchor=(1,1)
    ● Add grid to the plot using matplotlib grid() function.
    ● Show the plot using the show() function of matplotlib.

```python
sns.set(rc={'figure.figsize':(10,6)})
sns.set_palette('hls')
sns.set_style('darkgrid')
sns.set_context('paper', font_scale = 2,rc={'grid.linewidth': 0.5})
sns.swarmplot(data=df,x = 'Country', y='Life_Expectancy',hue='Year')
plt.title('Country wise Life Expectancy')
plt.xlabel('Country')
plt.ylabel('Country Wise Life Expectancy based on year')
plt.legend(ncol=2,fontsize=12,markerscale=0.8,bbox_to_anchor=(1,1))
plt.grid()
plt.show()
```

> # 3. 
    Create a Displot of 'Displot of Spending on Healthcare (in USD)'
