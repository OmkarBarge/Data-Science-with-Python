## 2.Plot the bird_2.jpg image in gray scale with Pillow and matplotlib using cmap='gray',and perform the necessary imports.

```python
img2 = Image.open('/content/bird_2.jpg')
gray_img2 = img2.convert('L')
plt.imshow(gray_img2,cmap='gray')
plt.show()
```

## 3.Plot the bird_2.jpg image with Pillow and matplotlib using cmap='inferno' and hide the axis by using axis()function,and perform the necessary imports.

```python
img2 = Image.open('/content/bird_2.jpg')
inferno_img2 = img2.convert('L')
plt.imshow(inferno_img2,cmap='inferno')
plt.axis('off')
plt.show()
```

## 4.Plot multiple images using subplot and apply the below customization<br>Plot the bird_1.jpg image using Pillow and matplotlib in row1,col1,and perform the necessary imports

```python
fig,(ax1,ax2) = plt.subplots(1,2,figsize=(12,10))

i1 = Image.open('/content/bird_1.jpg')
ax1.imshow(i1)

i2 = Image.open('/content/bird_2.jpg')
ax2.imshow(i2)

plt.show()
```

## Plot the bird_1.jpg image with Pillow and matplotlib using cmap='binary' in row1, col2, and perform the necessary imports.
