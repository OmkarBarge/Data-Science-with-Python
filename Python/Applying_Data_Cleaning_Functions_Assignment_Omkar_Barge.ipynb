```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/matches.csv')
```

```python
df.head()
```

```python
# 1.Find out the number of venues

venue_count = df['venue'].nunique()

print(f'There are {venue_count} venue')
```

```python
# 2.Find out the number of times each team emerged as the winner.


winner_count = pd.DataFrame(df['winner'].value_counts())

winner_count
```

```python
# 3.Find out the times the toss winner chose bat and the number of times the winner chose field.

pd.DataFrame(df.groupby('toss_winner')['toss_decision'].value_counts())
```

```python
# 4.Replace the toss_decision value with bowl_first where toss_decision is field.

df.loc[df['toss_decision'] == 'field' , 'toss_decision'] = 'bowl_first'
```

```python
df
```

```python
# 5.Sort the dataset in descending order of win by runs and print the column win_by_runs.


df.sort_values(by=['win_by_runs'], ascending=False,inplace=True)
df['win_by_runs']
```

```python
# 6.Find the date where the winner is Royal Challengers Bangalore with runs more than 30.

df['date'][(df['winner'] == 'Royal Challengers Bangalore') & (df['win_by_runs'] > 30)]
```

```python
# 7.Find the max win by runs and win by wickets using apply function.

df[['win_by_runs','win_by_wickets']].apply(np.max)
```
